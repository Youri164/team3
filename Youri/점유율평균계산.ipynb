```python
import pandas as pd
import numpy as np
import os
```

```python
ryu = pd.read_csv("류현진출전일점유율.csv")
ryu.head()
```

```python
# '구분'이 NaN이 아닌 값들의 '점유율' 평균 계산
non_nan_mean = ryu.loc[ryu['구분'].notna(), '점유율'].mean()

# '구분'이 NaN인 값들의 '점유율' 평균 계산
nan_mean = ryu.loc[ryu['구분'].isna(), '점유율'].mean()

print(f"'구분'이 NaN이 아닌 값들의 '점유율' 평균: {non_nan_mean:.2f}")
print(f"'구분'이 NaN인 값들의 '점유율' 평균: {nan_mean:.2f}")
```

```python
# 다른 폴더의 csv 파일 이용 (주피터노트북을 이용할하여 작업할 경우)

# 현재 작업 디렉토리 확인
script_dir = os.getcwd()  # 현재 작업 디렉토리
file_path = os.path.join(script_dir,'..','kwon','dc24_ratio.csv')
file_path = os.path.abspath(file_path)

# CSV 파일을 읽기

dc24 = pd.read_csv(file_path)

# data.info()
dc24.head()
```

```python
# 구장별 관중점유율
avg_occupancy_24 = dc24.groupby('구장_x')['점유율'].mean().reset_index()

print(avg_occupancy_24)
```

```python
avg_occupancy_24.to_csv('24년도점유율.csv', index=False, encoding='utf-8-sig')
```
