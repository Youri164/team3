```python
import pandas as pd
import numpy as np
```

```python
dc = pd.read_csv("daily_crowd_data.csv")
dc.head()
```

```python
ryu = pd.read_csv("ryu2024de.csv")
ryu.head()
```

```python
# "한화"가 포함된 행들만 필터링
hdc = dc[(dc['홈'] == '한화') | (dc['방문'] == '한화')]

# 결과 출력
print(hdc)
```

```python
# 류현진 상세지표 데이터 열을 문자열로 변환하고, '.'을 '/'로 대체
ryu['3월'] = ryu['3월'].apply(lambda x: str(x).replace('.', '/'))

# 결과 출력
print(ryu)
```
