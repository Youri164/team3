■ 정규시즌 

1. 정규리그 일정
   1) 일정 : 3월 20일 ~ 10월 31일
   3) 총 226일 (32주) / 주당 4.5경기 
  
2. 경기방식
   1) 단일 리그제 
      - 총 10개 구단      
      - 구단별 경기수 : 144 경기 = 9개 구단 * 16 차전 
      - 리그 총 경기수 : 720경기 = 10개 구단 * 144경기 / 2 
      - 홈경기와 방문경기는 50:50의 비율
   2) 연장전은 12회까지, 승패를 가리지 못하면 무승부로 한다

3. 선수관련
   1) 선수정원 : 최대 65명
   2) 외국인선수 : 3명 이내
   3) 현역선수 등록 : 28명(경기 참여 인원) 
      투수 - 14명
      포수 - 2명
      내야수 - 6명
      외야수 - 6명

4. 기록관련 (개인상 결정 최소기준)
   1) 타자 : 소속팀이 한 시즌에 치른 경기수의 3.1배 이상의 타석수
      예) 144경기 * 3.1 = 446.4 타석 이상
   2) 투수 : 소속팀이 한 시즌에 치른 경기수와 같은 수 이상의 횟수(이닝)
      예) 경기수 < 이닝수
